# Social Media vs Online Class

```python
df_so = df1[['social_media_time','avg_classtime_data','physical_problem_during_class']]
df_so.head(2)
```

```python
df_so_sum1 = pd.DataFrame(df_so)
df_so_sum1['time_new'] = np.sum(df_so[['social_media_time','avg_classtime_data']], axis=1)
df_so_sum1.tail()
```

```python
df_so_sum = df_so_sum1.groupby(['physical_problem_during_class'])[['social_media_time','avg_classtime_data']].mean()
df_so_sum.head()
```

So we can see, average class timing of both types of students are exactly the same, however there is a significant difference of almost an hour on average of social media usage. Those who faced physical issues use more social media than those who don't face issues.

```python
df_so_sum.plot.barh(stacked=False);
```

# Association Rule with sickness

```python
df_sickness = pd.read_csv('C:/Users\Asif\Desktop\Data Mining Assignment\Assignment_association_rule.csv')
df_sickness.head()
```

### Frequency Table

#### Table Preparation 

```python
df_sickness1 = df_sickness[['device_used_for_online_class','Headache','See_dimly','Hearing_impairment','Eyes_problem','Mentally_disturbed','Neck_pain','Backpains']]
df_sickness1
```

#### Frequency Table with Count

```python
df_sickness2 = df_sickness1.groupby(['device_used_for_online_class'])[['device_used_for_online_class','Headache','See_dimly','Hearing_impairment','Eyes_problem','Mentally_disturbed','Neck_pain','Backpains']].sum()
df_sickness2
```

### Applying Association Rule

```python
df_sickness3 = df_sickness[['Headache','See_dimly','Hearing_impairment','Eyes_problem','Mentally_disturbed','Neck_pain','Backpains']]
df_sickness3.head()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_itemsets = apriori(df_sickness3, min_support=0.02, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

Observations:

We can observe highest confidence for "Hearing Impairment" to "Headache" and "Eye Problem" to "Headache". Also "See dimly" to "Headache" has high confidence level.

A few rules with a high lift value, which means that it occurs more frequently than would be expected given the number of transaction and product combinations
Most of the places the confidence is high as well.

```python
rules[ (rules['lift'] >= 14) &
       (rules['confidence'] >= 0.2) ]
```

#### We can observe, that "Neck_pain" has highest confidence and lift with "Headache", "Backpains", "Neck_pain" with "See_dimly, Backpains", "Neck_pain, See_dimly" with "Headache, Backpains" , "Neck_pain, Headache"	with "See_dimly, Backpains", "Neck_pain" with "Headache, See_dimly, Backpains"
